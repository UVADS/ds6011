## Simulate some data and statistics

Supports slides: useful_statistics_and_error_analysis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(400,500,100)
x
```

```python
x.shape
```

```python
x.mean()
```

```python
x.std()
```

```python
y = x + np.random.normal(0,50,100)
y[10] = 200
y[80] = -500
y[90] = 1000
```

```python
plt.scatter(x,y)
plt.grid()
plt.xlabel('PTH (actual)')
plt.ylabel('PTH (predicted)')
```

```python
errors = y - x
```

```python
plt.hist(errors, 20)
plt.grid()
plt.xlabel('error')
plt.ylabel('frequency')
```

```python
abs_errors = np.abs(errors)
plt.plot(np.sort(abs_errors), np.linspace(0, 1, len(abs_errors), endpoint=False))
plt.grid()
plt.xlabel('absolute error')
plt.ylabel('cumulative density')
```

```python
print('mean:', round(abs_errors.mean(),2))
print('median:', round(np.median(abs_errors),2))
print('75%ile:', round(np.percentile(abs_errors, 75),2))
print('90%ile:', round(np.percentile(abs_errors, 90 ),2))
```
